```python
import matplotlib.pyplot as plt
import seaborn as srn
import pandas as pd
```

```python
# grafico de barras

linguagens = ['C', 'C++', 'Java', 'Python','PHO']
estudantes = [23,17,35,29,12]

plt.bar(linguagens,estudantes)
plt.xlabel('estudantes')
plt.ylabel('linguagens')
plt.title('Número de alunos matriculados por linguagem')
plt.show()
```

```python
# gráfico de barras na horizontal (colocar um h no bar)

linguagens = ['C', 'C++', 'Java', 'Python','PHO']
estudantes = [23,17,35,29,12]

plt.barh(linguagens,estudantes)
plt.xlabel('estudantes')
plt.ylabel('linguagens')
plt.title('Número de alunos matriculados por linguagem')
plt.show()
```

```python
base = pd.read_csv('penguins_size.csv')
```

```python
base
```

```python
# vamos fazer um histograma com seaborn, para ver a distrbuiçao dos dados
# a média está maior que a moda(valor que mais se repete), é uma assimetria positiva

srn.histplot(base['body_mass_g'],bins=10,color='blue')
```

```python
# A linha do meio é considerada a mediana, não está no meio do bloco.
# Portanto há um assimetria.
srn.boxplot(base.body_mass_g).set_title('Peso')
```

```python
# Histograma do tamanho da nadadeira dos penguins (flipper_length)
srn.histplot(base.loc[:,'flipper_length_mm'],bins=10,color='blue')
```
